# Water Transportation (Money)

```python
driver = gs.Chrome()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# I Initialize the WebDriver
driver.get("https://azstat.gov.az/statHtml/statHtml.do?orgId=994&tblId=DT_BJ_046&vw_cd=MT_ATITLE&list_id=&scrId=&seqNo=&language=az&obj_var_id=&conn_path=I2&path=&userId=")

try:
    # Works for: Wait for the table to be visible
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.ID, "popMode"))
    )

    # Works for: Extract the HTML content of the table
    table_html = element.get_attribute('outerHTML')
finally:
    driver.quit()

# I Use pandas to parse the HTML table
df = pd.read_html(table_html)  # Returns a list of DataFrames
if df:
    df_water_money = df[0]
    print(df_water_money)

```

# Merging dataframes

```python
import pandas as pd
from functools import reduce

# Assuming df1, df2, df3, df4, df5, df6 are the dataframes
# List of all dataframes
dataframes = [df_water_money, df_water_load, df_air_money, df_air_load, df_railw_tr_money, df_rail_tr_load]

# I Find common year columns across all dataframes
common_years = list(set(df_water_money.columns) & set(df_water_load.columns) & set(df_air_money.columns) & set(df_air_load.columns) & set(df_railw_tr_money.columns) & set(df_rail_tr_load.columns))

# I Use functools.reduce to merge all dataframes on 'Göstərici' and common years
screaped_data_team_38 = reduce(lambda left, right: pd.merge(left, right, on=["Göstərici"] + common_years, how="outer"), dataframes)

print(screaped_data_team_38)
```

```python
screaped_data_team_38.to_csv("screaped_data_team_38.csv", index=False)  #I Set index=False to avoid saving the index column
```
