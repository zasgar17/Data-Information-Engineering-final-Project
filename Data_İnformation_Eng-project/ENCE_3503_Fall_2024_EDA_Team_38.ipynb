```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/scraped_data_team_38 (2).csv')
df.head()
```

```python
# Step 3: Data Cleaning

df = df.drop(columns=['1995_x', '1995_y'], errors='ignore')
```

```python
# here I am filling in the missing values in each column by using the column's mean
df = df.apply(lambda x: x.fillna(x.mean()) if x.dtype in ['float64', 'int64'] else x)
```

```python
df
```

```python
# Step 4: Exploratory Data Analysis
# now I am generating summary statistics for the dataset to analyze its key metrics
print("Summary Statistics:")
print(df.describe())
```

```python
# lets create visualizations to better understand the data
def plot_trends(df, category_col, start_year=1995, end_year=2023):
    """Visualize trends across years for each category."""
    years = [str(year) for year in range(start_year, end_year + 1) if str(year) in df.columns]
    plt.figure(figsize=(12, 8))
    for category in df[category_col].unique():
        subset = df[df[category_col] == category]
        plt.plot(years, subset[years].mean(), label=category)
    plt.xlabel("Year")
    plt.ylabel("Values")
    plt.title(f"Trends for {category_col} Across Years")
    plt.legend()
    plt.show()
```

```python
# here I am calling the visualization function to display the data graphically
if 'göstərici' in df.columns:
    plot_trends(df, 'göstərici', 1995, 2023)
else:
    print("Column 'göstərici' not found in the DataFrame.")
```

```python
# here correlations between numerical columns is shown
plt.figure(figsize=(15, 13))
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 5: Save the cleaned data
processed_file_path = 'processed_data_team_38.csv'
df.to_csv(processed_file_path, index=False)
print(f"Cleaned data saved to {processed_file_path}")
```
